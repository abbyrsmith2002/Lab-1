```python
import polars as pl
```

```python
Pitching = pl.read_csv('./Data/lahman_1871-2023_csv/Pitching.csv')
```

```python
Pitching.head()
```

```python
Batting = pl.read_csv('./Data/lahman_1871-2023_csv/Batting.csv')
Batting.head()
```

```python
League_Wide_ERA = (Pitching
                     .select([pl.col('IPouts'), pl.col('yearID'), pl.col('R'), pl.col('lgID')])
                     .filter((pl.col('yearID') >= 1946))
                     .group_by(pl.col('yearID'), pl.col('lgID'))
                     .agg([pl.col('R').sum().alias('total_runs'),
                           pl.col('IPouts').sum().alias('total_ipouts')
                          ])
                     .with_columns(League_ERA = pl.col('total_runs')/pl.col('total_ipouts')*27)
                    )
League_Wide_ERA
```

```python
League_Wide_ERA.head()
```

```python
League_Wide_ERA.write_csv('League_Wide_ERA.csv')
```

```python
DH = (Pitching
      .select([pl.col('IPouts'), pl.col('yearID'), pl.col('R'), pl.col('lgID')])
      .with_columns(
          (
              pl.when(pl.col('yearID') < 1972)
              .then(pl.lit('Neither'))
              .when(pl.col('yearID') < 2019)
              .then(pl.lit('Only AL'))
              .when(pl.col('yearID') == 2020)
              .then(pl.lit('Both'))
              .when(pl.col('yearID') == 2021)
              .then(pl.lit('Only AL'))
              .when(pl.col('yearID') >= 2022)
              .then(pl.lit('Both'))
              .otherwise(pl.lit(None))
              .alias('DH')
          )
      )
)

DH
```

```python
DH.head()
```

```python
DH.write_csv('DH.csv')
```

```python
Batting = pl.read_csv('./Data/lahman_1871-2023_csv/Batting.csv')
```

```python
Batting.head()
```

```python
Slugging_Pct = (Batting
                     .select([pl.col('H'), pl.col('2B'), pl.col('3B'), pl.col('HR'), pl.col('AB'), pl.col('yearID'), pl.col('lgID')])
                     .filter((pl.col('yearID') >= 1946))
                     .group_by(pl.col('yearID'), pl.col('lgID'))
                     .agg([pl.col('2B').sum().alias('total_doubles'),
                           pl.col('3B').sum().alias('total_triples'),
                                 pl.col('HR').sum().alias('total_homeruns'),
                                       pl.col('H').sum().alias('total_Hits'),
                           pl.col('AB').sum().alias('total_AB')
                          ])
                .with_columns(total_singles = pl.col('total_Hits') - pl.col('total_doubles') - pl.col('total_triples')
                                                  - pl.col('total_homeruns'))
                     .with_columns(Slugging = (pl.col('total_singles')+(pl.col('total_doubles')*2)+(pl.col('total_triples')*3)+
                                   (pl.col('total_homeruns')*4))/(pl.col('total_AB'))
                    ))
Slugging_Pct
```

```python
Slugging_Pct.write_csv('Slugging_Pct.csv')
```

```python
Slugging_Pct.head()
```

```python
import polars as pl
import plotnine as pn
```

```python
pn.options.figure_size = (6, 4)
```

```python
pivot_ERA = (League_Wide_ERA
 .pivot('lgID', 
        index = 'yearID', 
        values = "League_ERA"
       )
)
pivot_ERA
```

```python
diff_ERA = pivot_ERA.with_columns((pl.col("AL")-pl.col("NL")).alias("AL-NL"))
diff_ERA
```

```python
diff_ERA = (diff_ERA
 .unpivot(["AL","NL", "AL-NL"], 
         index='yearID', 
         variable_name='League', 
         value_name='League_Wide_ERA')
)
```

```python
from plotnine import ggplot, geom_bar, geom_line, aes, geom_vline, annotate, theme

(ggplot(data = diff_ERA)
 + geom_line(aes(x = "yearID",                    
                y = "League_Wide_ERA",
                color = "League")
            ) 
 + geom_vline(xintercept=1973, linetype='dashed', color='orange')
 + annotate('text', x=diff_ERA['yearID'].min()+4, y=diff_ERA['League_Wide_ERA'].min()+1, 
            label='DH introduced to the AL in 1973', ha='left', va='bottom', size=12, color='black')
  + theme(figure_size=(12, 6)) 
            )
```

```python
Slugging_Pct.head()
```

```python
Slugging_Perct = (Slugging_Pct
 .pivot('lgID', 
        index = 'yearID', 
        values = "Slugging"
       )
                  .with_columns((pl.col("AL")-pl.col("NL")).alias("AL-NL"))
                  .unpivot(["AL","NL", "AL-NL"], 
         index='yearID', 
         variable_name='League', 
         value_name='Slugging')
)
Slugging_Perct
```

```python
from plotnine import ggplot, geom_bar, geom_line, aes, geom_vline, annotate, theme

(ggplot(data = Slugging_Perct)
 + geom_line(aes(x = "yearID",                    
                y = "Slugging",
                color = "League")
            ) 
 + geom_vline(xintercept=1973, linetype='dashed', color='orange')
 + annotate('text', x=Slugging_Perct['yearID'].min(), y=Slugging_Perct['Slugging'].min()+0.2, 
            label='DH introduced to the AL in 1973', ha='left', va='bottom', size=11, color='black')
 + theme(figure_size=(10, 6)) 
            )
```
